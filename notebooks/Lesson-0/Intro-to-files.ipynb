#How to Read and Write Files

##Binary

Computers store information in many different ways, but they all are based on binary. Binary is a number system composed of only two symbols one and zero. What one means depends on the storage medium, it can mean a high voltage, a magnetic pointing up, a circuit being on or many other things, zero is then the opposite, a low voltage, a magnet pointing down or a circuit being off.

Since binary is a number system you can translate decimal numbers (the number system most people use) into binary. This is a simple process, you simply add up the values of every digit just like you do with decimal numbers. For example lets break up one hundred and twenty three:
